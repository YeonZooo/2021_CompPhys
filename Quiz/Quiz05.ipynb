```python
# 1-1
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('./quiz5-1.txt',dtype=None,delimiter='    ')
t = data[:,0]; y = data[:,1]

pfit1 = np.polyfit(t,y,1)
line1 = np.poly1d(pfit1)

pfit2 = np.polyfit(t,y,2)
line2 = np.poly1d(pfit2)

pfit4 = np.polyfit(t,y,4)
line4 = np.poly1d(pfit4)

tt = np.linspace(0,1.4,100)

plt.plot(t,y,'*',label='data')
plt.plot(tt,line1(tt),'-',label='fitting1')
plt.plot(tt,line2(tt),'-',label='fitting2')
plt.plot(tt,line4(tt),'-',label='fitting4')
plt.xlabel('t [sec]')
plt.ylabel('y [m]')
plt.legend()
plt.show()

# 1-2
"""
1-1의 plot를 보면, data는 4차 함수와 가장 큰 일치를 보인다.
하지만 더 많은 데이터를 측정할 경우에는 물리적 직관에 의해 그래프는 포물선을 그릴 것을 얘측할 수 있고,
이는 2차 함수와 가장 잘 맞아 떨어질 것이다.
"""

# 1-3
# pfit2 = [-0.5*g, v0, y0]
print(f"""
v0 = {round(pfit2[1],2)} m/s 
g  = {round(-2*pfit2[0],2)} m/s^2""")
```

```python
# 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

data = np.genfromtxt('./quiz5-1.txt',dtype=None,delimiter='    ')
t = data[:,0]; y = data[:,1]

def fitfun(t,a,b):
    return a*t*t + b*t    # -0.5*g*t**2 + v_0*t

parm, cov = curve_fit(fitfun, t, y)

tt = np.linspace(0,1.4,100)
yy = fitfun(tt,parm[0],parm[1])

plt.plot(t,y,'*',label='data')
plt.plot(tt,yy,'-',label='curve_fit')
plt.xlabel('t [sec]')
plt.ylabel('y [m]')
plt.legend()
plt.show()

print(f"""
v0 = {round(parm[1],2)} m/s 
g  = {round(-2*parm[0],2)} m/s^2""")
```

```python
# 3
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

data = np.genfromtxt('./quiz5-3.txt',dtype=None,delimiter=' ')
x = data[:,0]; y = data[:,1]

def gauss(x,a,b,c):
    return a*np.exp(-(x-b)**2/c**2)

def fitfun(x,a1,b1,c1,a2,b2,c2,a3,b3,c3):
    return gauss(x,a1,b1,c1) + gauss(x,a2,b2,c2) + gauss(x,a3,b3,c3)

parm, cov = curve_fit(fitfun, x, y, p0=[1.4,3,0.3,6.5,5,1,6,7,2])

xx = np.linspace(0,10,100)
yy = fitfun(xx,parm[0],parm[1],parm[2],parm[3],parm[4],parm[5],parm[6],parm[7],parm[8])

plt.plot(x,y,'*',label='data')
plt.plot(xx,yy,'k-',label='curve_fit')
plt.plot(xx,gauss(xx,parm[0],parm[1],parm[2]),'r-',label='gauss1')
plt.plot(xx,gauss(xx,parm[3],parm[4],parm[5]),'y-',label='gauss2')
plt.plot(xx,gauss(xx,parm[6],parm[7],parm[8]),'g-',label='gauss3')
plt.legend()
plt.show()

print(f"""
a1 = {round(parm[0],1)}, b1 = {round(parm[1],1)}, c1 = {round(parm[2],1)}
a2 = {round(parm[3],1)}, b2 = {round(parm[4],1)}, c2 = {round(parm[5],1)}
a3 = {round(parm[6],1)}, b3 = {round(parm[7],1)}, c3 = {round(parm[8],1)}""")
```
