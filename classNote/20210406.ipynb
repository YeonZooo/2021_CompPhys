```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import *
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,6,3,6,4])

pfit = np.polyfit(x,y,1)
line = np.poly1d(pfit)

xx = np.linspace(0,6,50)
yy = line(xx)

plt.plot(xx,yy)
plt.plot(x,y,'*')
plt.show()
```

```python
from scipy.optimize import curve_fit

def f(x,m,b):
    return m*x + b

parm, cvar = curve_fit(f,x,y)
yy1 = f(xx,parm[0],parm[1])

plt.plot(x,y,'*')    # original data
plt.plot(xx,yy)    # linear fit
plt.plot(xx,yy1,'--')    # second linear fit
plt.show()
```

```python
x = np.array([-2,1,4,-1,3,-4])
y = np.array([-1,2,59,4,24,-53])

pfit = np.polyfit(x,y,3)
line = np.poly1d(pfit)

xx = np.linspace(-5,5,50)
yy = line(xx)

plt.plot(xx,yy)
plt.plot(x,y,'*')
plt.show()
```

```python
data = np.loadtxt('./data/Data2.txt',dtype=None,delimiter=',')

x = data[0]; y = data[1]

def f(x,A,B,C):
    return A*np.sin(B*x + C)

parm, cvar = curve_fit(f,x,y, p0=[0.5,2,1.5])

xx = np.linspace(0,10,100)
yy1 = f(xx,parm[0],parm[1],parm[2])

plt.plot(xx,yy1)
plt.plot(x,y,'*')
plt.show()
```

```python
data = np.loadtxt('./data/oildata.dat',dtype=None,delimiter=',')

t = data[0]; y = data[1]

def f(t,v0,tau):
    return v0*(1-np.exp(-t/tau))

parm, cvar = curve_fit(f,t,y)

tt = np.linspace(0,1,100)
yy = f(tt,parm[0],parm[1])

plt.plot(tt,yy)
plt.plot(t,y,'*')
plt.show()

print(parm)
```
